### Label data to prepare for supervised learning

```python
# drop rows with missing values
print('initial shape (17424, 44)')
df_290_red_samp.dropna(inplace=True)
print(f'new shape: {df_290_red_samp.shape}')
```

```python
# Create the scatter plot using Plotly Express
fig = px.scatter(df_290_red_samp,
                 x=df_290_red_samp['Timestamp'],
                 y=df_290_red_samp['Alt(m)'])

# Update the layout with title and axis labels
fig.update_layout(
        title='Altitude over Time',
        xaxis_title='Timestamp',
        yaxis_title='Altitude (m)',
        coloraxis_colorbar=dict(title='Cluster Label')
    )

# Show the Plotly figure
fig.show()
```

```python
# Initialize 'on_lift' column to 0
df_290_red_samp['on_lift'] = 0
```

```python
# Manually noted timestamps for lift rides
lift_rides = [
    ('2024-01-27 09:02:26', '2024-01-27 09:03:46'),
    ('2024-01-27 09:08:19', '2024-01-27 09:12:16'),
    ('2024-01-27 09:15:43', '2024-01-27 09:19:56'),
    ('2024-01-27 09:23:06', '2024-01-27 09:29:16'),
    ('2024-01-27 09:34:33', '2024-01-27 09:42:44'),
    ('2024-01-27 09:45:12', '2024-01-27 09:51:42'),
    ('2024-01-27 09:54:50', '2024-01-27 09:58:15'),
    ('2024-01-27 10:03:01', '2024-01-27 10:06:45'),
    ('2024-01-27 10:07:41', '2024-01-27 10:09:36'),
    ('2024-01-27 10:49:38', '2024-01-27 10:51:41'),
    ('2024-01-27 10:58:42', '2024-01-27 11:03:44'),
    ('2024-01-27 11:10:09', '2024-01-27 11:12:35'),
    ('2024-01-27 11:16:20', '2024-01-27 11:18:55'),
    ('2024-01-27 11:27:13', '2024-01-27 11:32:23'),
    ('2024-01-27 13:51:51', '2024-01-27 13:52:49'),




]
```

```python
df_290_red_samp['Timestamp'] = pd.to_datetime(df_290_red_samp['Timestamp'])
```

```python
# update 'on_lift' column
for start, end in lift_rides:
    start_time = pd.to_datetime(start)
    end_time = pd.to_datetime(end)
    df_290_red_samp.loc[(df_290_red_samp['Timestamp'] >= start_time) & (df_290_red_samp['Timestamp'] <= end_time), 'on_lift'] = 1
```

```python
#def plot_results():
    # Plot for Altitude over Time, colored by 'on_lift' status
plt.figure(figsize=(15, 10))

    # Plot points where 'on_lift' is 1 with one color
plt.scatter(df_290_red_samp.loc[df_290_red_samp['on_lift'] == 1, 'Timestamp'],
                df_290_red_samp.loc[df_290_red_samp['on_lift'] == 1, 'Alt(m)'],
                c='red',
                label='On the lift',
                marker='o')

    # Plot points where 'on_lift' is 0 with another color
plt.scatter(df_290_red_samp.loc[df_290_red_samp['on_lift'] == 0, 'Timestamp'],
                df_290_red_samp.loc[df_290_red_samp['on_lift'] == 0, 'Alt(m)'],
                c='blue',
                label='Not on the lift',
                marker='o')

plt.title('Labeled data df_290')
plt.xlabel('Timestamp')
plt.ylabel('Altitude (m)')
plt.legend()
plt.show()
```
