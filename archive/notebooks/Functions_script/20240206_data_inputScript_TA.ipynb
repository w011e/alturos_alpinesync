```python

def fetch_and_process_data(file_path, chunk_size=60, wait_time=2):
    # Load the dataset
    df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/Sample_data/v5_20240204_150730_57m.csv')

    # Calculate the total number of chunks
    total_chunks = len(df) // chunk_size

    all_calculations = []  # Initialize an empty list to store all calculations

    for i in range(total_chunks):
        # Fetch a chunk of 60 observations
        start_index = i * chunk_size
        end_index = start_index + chunk_size
        data_chunk = df.iloc[start_index:end_index]

        if len(data_chunk) > 0:
            # Calculate the mean and median of the specified features
            mean_a = data_chunk[['accelX(g)', 'accelY(g)', 'accelZ(g)']].mean(axis=1)
            median_a = data_chunk[['accelX(g)', 'accelY(g)', 'accelZ(g)']].median(axis=1)

            # Add the calculated mean and median to the data_chunk as new columns
            data_chunk = data_chunk.assign(mean_a=mean_a, median_a=median_a)

            # Append the processed chunk to the list of all calculations
            all_calculations.append(data_chunk)

            # Print the first few rows of the processed chunk for demonstration
            print(f"Processed chunk {i+1} with new features 'mean_a' and 'median_a':")
            print(data_chunk[['accelX(g)', 'accelY(g)', 'accelZ(g)', 'mean_a', 'median_a']].head())
        else:
            print(f"No more data to process in chunk {i+1}.")

        # Wait for 60 seconds before processing the next chunk
        time.sleep(wait_time)

# Example usage
file_path = 'content/drive/MyDrive/Colab Notebooks/Sample_data/v5_20240204_150730_57m.csv'
fetch_and_process_data(file_path)
```
