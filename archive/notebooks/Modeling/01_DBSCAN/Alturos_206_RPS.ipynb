```python
from sklearn.cluster import DBSCAN
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
import pandas as pd
import numpy as np
```

```python
#Select relevant features
features = ['Course','Speed(m/s)']
X = df_resampled[features]
```

```python
#Pipeline for scaling and clustering
pipeline = Pipeline([
    ('scaler', StandardScaler()),
    ('dbscan', DBSCAN(
        eps=0.25,
        min_samples=50))
])

#Fit the pipeline
clusters = pipeline.fit_predict(X)

#Add the cluster labels to your dataframe
df_resampled['Cluster'] = clusters
# Increment cluster labels by 1 to treat noise points as a separate cluster
#df_resampled['Cluster'] += 1
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

#2D Plot for Altitude over Time
plt.figure(figsize=(12, 6))
plt.scatter(df_resampled['Timestamp'],
            df_resampled['Alt(m)'],
            c=df_resampled['Cluster'],
            cmap='viridis',
            marker='o')

plt.title('Altitude over Time by Cluster')
plt.xlabel('Timestamp')
plt.ylabel('Altitude (m)')
plt.colorbar(label='Cluster Label')
plt.show()


#2D Plot for Speed over Time
plt.figure(figsize=(12, 6))
plt.scatter(df_resampled['Timestamp'],
            df_resampled['Speed(m/s)'],
            c=df_resampled['Cluster'],
            cmap='viridis',
            marker='o')

plt.title('Speed over Time by Cluster')
plt.xlabel('Timestamp')
plt.ylabel('Speed (m/s)')
plt.colorbar(label='Cluster Label')
plt.show()
```
